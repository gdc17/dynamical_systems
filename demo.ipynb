```python
import numpy as np
import matplotlib.pyplot as plt
```

Investigate the logistic equation $x_{n+1}=rx_n(1-x_n)$.

```python
K = 15000 # number of trajectories in ensemble
N = 100 # number of time-steps
x = np.zeros((N,K))
x[0,:] = np.random.uniform(0,1,size=K) # randomly distirbuted initial conditions

# Pick your parameter
r = 3.7

for i in range(N-1):
    x[i+1,:] = r*x[i,:]*(1-x[i,:]);

# Plot historgram of ensemble at n=100   
plt.hist(x[-2,:], bins = 50, density = True);
plt.xlim([0,1]);

# Plot time series of a single trajectory
plt.figure()
plt.plot(x[:,0]);
plt.ylim([0,1]);
```
